## 4. ‚öôÔ∏è EXECU√á√ÉO PERSONALIZADA

Como configurar o agente para necessidades espec√≠ficas:

```python
# ==========================================
# OP√á√ÉO 3: EXECU√á√ÉO PERSONALIZADA
# ==========================================

print("‚öôÔ∏è CONFIGURA√á√ÉO PERSONALIZADA DO AGENTE")
print("="*50)

# Configura√ß√£o personalizada
config_personalizada = ConfiguracaoProjeto(
    metrica_otimizar='r2',  # Mudar m√©trica para R¬≤
    n_iter_otimizacao=20,   # Menos itera√ß√µes para ser mais r√°pido
    cv_folds=5,             # Menos folds para ser mais r√°pido
    diretorio_cache='cache_personalizado'  # Cache separado
)

# Criar agente personalizado
agente_personalizado = AgenteMLPetrobras(config_personalizada)

# Executar com algoritmos espec√≠ficos
algoritmos_selecionados = ['linear', 'ridge', 'random_forest', 'xgboost']

print(f"üéØ M√©trica de otimiza√ß√£o: {config_personalizada.metrica_otimizar}")
print(f"üîÑ Itera√ß√µes de otimiza√ß√£o: {config_personalizada.n_iter_otimizacao}")
print(f"üìä Folds de valida√ß√£o cruzada: {config_personalizada.cv_folds}")
print(f"üß™ Algoritmos selecionados: {algoritmos_selecionados}")

# Executar miss√£o personalizada
resultados_personalizados = agente_personalizado.executar_missao_completa(
    algoritmos_selecionados=algoritmos_selecionados,
    usar_cache=True,
    eda_completa=True
)

print("\n‚úÖ EXECU√á√ÉO PERSONALIZADA CONCLU√çDA!")
```

---
## 5. üíæ CARREGAMENTO E USO DE MODELOS SALVOS

Como carregar e usar os modelos treinados:

```python
# ==========================================
# CARREGAMENTO DE MODELOS SALVOS
# ==========================================

print("üíæ DEMONSTRA√á√ÉO DE CARREGAMENTO DE MODELOS")
print("="*50)

# Carregar modelo do agente completo
modelo_final = agente_completo.carregar_modelo_salvo()

if modelo_final:
    print("\nüîÆ FAZENDO PREDI√á√ïES COM O MODELO FINAL:")
    
    # Carregar dados para exemplo
    df = agente_completo.df_processado
    X = df.drop(agente_completo.target, axis=1)
    y = df[agente_completo.target]
    
    # Fazer predi√ß√µes em algumas amostras
    amostras_teste = X.head(5)
    predicoes = modelo_final.predict(amostras_teste)
    valores_reais = y.head(5)
    
    print("\nüìä COMPARA√á√ÉO: REAL vs PREDITO")
    for i in range(5):
        print(f"  Amostra {i+1}: Real={valores_reais.iloc[i]:.4f}, Predito={predicoes[i]:.4f}, Erro={abs(valores_reais.iloc[i] - predicoes[i]):.4f}")
    
    # Estat√≠sticas do modelo
    from sklearn.metrics import r2_score, mean_squared_error
    todas_predicoes = modelo_final.predict(X)
    r2_final = r2_score(y, todas_predicoes)
    rmse_final = np.sqrt(mean_squared_error(y, todas_predicoes))
    
    print(f"\nüìà M√âTRICAS FINAIS DO MODELO:")
    print(f"  R¬≤ (R-squared): {r2_final:.4f}")
    print(f"  RMSE: {rmse_final:.4f}")
else:
    print("‚ùå Nenhum modelo encontrado. Execute uma das op√ß√µes acima primeiro.")
```

## 7. üîß UTILIT√ÅRIOS E CACHE

Ferramentas para gerenciar cache e arquivos:

```python
# ==========================================
# GERENCIAMENTO DE CACHE E ARQUIVOS
# ==========================================

print("üîß UTILIT√ÅRIOS DE GERENCIAMENTO")
print("="*50)

# Verificar arquivos de cache
from pathlib import Path
import os

def verificar_cache():
    cache_dirs = ['cache_agente_ml', 'cache_personalizado']
    
    for cache_dir in cache_dirs:
        if Path(cache_dir).exists():
            cache_files = list(Path(cache_dir).glob("*.joblib"))
            print(f"üìÅ {cache_dir}: {len(cache_files)} arquivos")
            for f in cache_files:
                size_mb = f.stat().st_size / (1024*1024)
                print(f"   - {f.name}: {size_mb:.1f}MB")
        else:
            print(f"üìÅ {cache_dir}: n√£o existe")

verificar_cache()

# Verificar modelos salvos
print("\nüíæ MODELOS SALVOS:")
model_files = list(Path('.').glob("modelo_final_*.joblib"))
if model_files:
    for f in model_files:
        size_mb = f.stat().st_size / (1024*1024)
        print(f"   - {f.name}: {size_mb:.1f}MB")
else:
    print("   Nenhum modelo encontrado")

# Fun√ß√£o para limpar cache (se necess√°rio)
def limpar_cache_se_solicitado():
    resposta = input("\nüóëÔ∏è Deseja limpar o cache? (s/n): ")
    if resposta.lower() in ['s', 'sim', 'y', 'yes']:
        import shutil
        cache_dirs = ['cache_agente_ml', 'cache_personalizado']
        for cache_dir in cache_dirs:
            if Path(cache_dir).exists():
                shutil.rmtree(cache_dir)
                print(f"‚úÖ Cache {cache_dir} removido")
        print("‚úÖ Limpeza conclu√≠da!")
    else:
        print("Cache mantido.")

# Chamar fun√ß√£o (descomente se quiser usar)
# limpar_cache_se_solicitado()
```
